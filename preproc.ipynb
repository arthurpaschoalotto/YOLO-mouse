```python
# resetou index e excluiu os valores nulos Nan
df_train = df_train.reset_index(drop=True).dropna()
df_test = df_test.reset_index(drop=True).dropna()
```

```python
df_train['split'] = 'train'
df_test['split'] = 'test'

df_splits = pd.concat([df_train, df_test], ignore_index = True)
df_splits
```

```python
import PIL
import cv2

def resize_img(img):
    width = 224
    height = 224
    dim = (width, height)
    img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    return img
```

```python
for row in range(len(df_splits)):
    img = cv2.imread(df_splits['path'][row])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = resize_img(img)
    cv2.imwrite(r"C:\\Users\\Arthur\\Desktop\\Desafio\\data_proc\\" +str(df_splits['split'][row]) + "/" + str(row) + ".jpg", cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

    print(r"C:\\Users\\Arthur\\Desktop\\Desafio\\data_proc\\" +str(df_splits['split'][row]) + "/" + str(row)+".jpg")
```
